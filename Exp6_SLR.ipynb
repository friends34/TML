```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/ML(P)/salary_data.csv')
```

```python
df.head()
```

```python
df.plot(x="YearsExperience",y="Salary",style='o')
plt.xlabel('Experience')
plt.ylabel('Salaries')
plt.show()
```

```python
X=pd.DataFrame(df['YearsExperience'])
y=pd.DataFrame(df['Salary'])
```

STEP-1: Splitting training and testing data

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=1/3,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

STEP-2: Fitting SLR to training set

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
print(regressor.intercept_)
```

```python
print(regressor.coef_)
```

STEP-3: Predicting test and train result

```python
y_pred= regressor.predict(X_test)
x_pred= regressor.predict(X_train)
```

STEP-4: Visualize train data

```python
plt.scatter(X_train, y_train, color="green")
plt.plot(X_train, x_pred, color="red")
plt.title("Salary vs Experience (Training Dataset)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary(In Rupees)")
plt.show()
```

STEP-5: Visualize test data

```python
plt.scatter(X_test, y_test, color="blue")
plt.plot(X_train, x_pred, color="red")
plt.title("Salary vs Experience (Test Dataset)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary(In Rupees)")
plt.show()
```

STEP-6: Evaluating the algorithm

```python
from sklearn import metrics
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,y_pred))
print('Mean squared Error:',metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
